##### Her Videonun Başlık Uzunluğunu içeren yeni bir sütün ekleyin - Add a new column containing the Title Length of each Video 

```python
data["length"] = data["title"].apply(len)
```

##### Her bir videoda kullanılan etiket sayısını hesaplayarak, oluşan değerleri yeni bir sütun olarak ekleyin - Find out how many tags are in each video and add a new column of these values

Not: Her bir etiket "|" işaretiyle ayrılmaktadır.

```python
def tag(x):
    ls = x.split("|")
    return len(ls)
data["tag_count"] = data["tags"].apply(tag)
```

##### (Challenge) Videoları Beğeni Oranı en yüksek olandan en düşüğe doğru sıralayınız - Sort  videos from highest to lowest likes

```python
def like_dislike(likes,dislikes):
    lslike = []
    for key,val in list(likes.iteritems()):
        lslike.append(val)
    lsdislike = []
    for key,val in list(dislikes.iteritems()):
        lsdislike.append(val)
    likedislike = list(zip(lslike,lsdislike))
    res = []
    for like,dislike in likedislike:
        if like+dislike == 0:
            res.append(0)
        else:
            res.append( like / (dislike+like)) 
    return res
    print(res)
               
               
data["likes_dislikes"] = like_dislike(data["likes"],data["dislikes"])
```

```python
data.sort_values("likes_dislikes",ascending = False,inplace = True)
```
