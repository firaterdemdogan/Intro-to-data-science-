```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(1,11,2)
```

```python
y
```

```python
fig,axes =plt.subplots(2,1)
axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x**3)
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (4,4))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = "purple")
```

```python
fig,axes = plt.subplots(2,1,figsize=(8,6))

axes[0].plot(x,y,color = "red")
axes[0].set_title("First Axes")
axes[1].plot(x,x ** 0.5,color = "purple")
axes[1].set_title("Second Axes")
plt.tight_layout()
```

```python
fig.savefig("Figure1.png")
```

```python
fig = plt.figure(figsize = (8,6))
axes =  fig.add_axes([0,0,1,1])
axes.plot(x,x**0.5,color = "red",label = "X Karekk")
axes.plot(x,x**2,color = "yellow",label = "X Kare")
axes.plot(x,x**3,color = "#4055c2",label = "X Kp")
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,color = "red",linewidth = 2,marker = "o",markersize = 10,markerfacecolor = "black",markeredgecolor= "yellow",markeredgewidth = 3     )
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",linewidth = 2,marker = "o",markersize = 10,markerfacecolor = "black",markeredgecolor= "yellow",markeredgewidth = 3 )

axes.set_xlim(0,10)
axes.set_ylim(0,50)
```
